```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Construir o caminho para o arquivo CSV
file_path = r"d:\Users\Administrator\Documents\Github\014-Pos-Graduacao-Engenharia-de-Dados\data\caso_full.csv.gz"
```

```python
# Carregar o arquivo CSV
df = pd.read_csv(file_path)
```

```python
# Visualizando os 5 primeiros dados
df.head()
```

```python
# Gerando a estatistica dos dados
df.describe()
```

```python
# Números de casos confirmados por estado
df_grouped = df.groupby(['state'])['new_confirmed'].sum().reset_index()
print(df_grouped)
```

```python
# 1. Gráfico de barras - Número de casos por estado
cases_by_state = df.groupby("state")["new_confirmed"].sum()
plt.figure(figsize=(10, 6))
cases_by_state.plot(kind="bar")
plt.title("Número de casos por estado")
plt.xlabel("Estado")
plt.ylabel("Número de casos")
plt.show()
```

```python
# 2. Gráfico de barras - Número de casos por Mês
# Converter a coluna de data para o tipo datetime, caso ainda não esteja
df['date'] = pd.to_datetime(df['date'])

# Extrair o mês de cada data e adicionar uma coluna "month" ao DataFrame
df['month'] = df['date'].dt.month

# Agrupamento por mês, somando os casos confirmados
df_grouped = df.groupby('month')['new_confirmed'].sum()

# Gerar o gráfico de barras
plt.bar(df_grouped.index, df_grouped.values)
plt.xlabel('Mês')
plt.ylabel('Total de Casos Confirmados')
plt.title('Quantidade de Casos Confirmados por Mês')
plt.xticks(df_grouped.index)  # Definir os rótulos do eixo x como os meses
plt.show()
```

```python
# 3. Gráfico de linhas - Evolução do número de casos ao longo do tempo
cases_over_time = df.groupby("date")["new_confirmed"].sum()
plt.figure(figsize=(10, 6))
cases_over_time.plot(kind="line")
plt.title("Evolução do número de casos de COVID-19 ao longo do tempo")
plt.xlabel("Data")
plt.ylabel("Número de casos")
plt.show()
```

```python
# 4. Gráfico apresenta o percentual de pessoas infectadas em relação à população estimada
import pandas as pd
import matplotlib.pyplot as plt

# Agrupar por estado, somando os casos confirmados e a população estimada
df_grouped = df.groupby('state').agg({'new_confirmed': 'sum', 'estimated_population': 'first'}).reset_index()

# Calcular o percentual de pessoas infectadas em relação à população estimada
df_grouped['percent_infected'] = (df_grouped['new_confirmed'] / df_grouped['estimated_population']) * 100

# Remover linhas com valores NaN na coluna "percent_infected"
df_grouped = df_grouped.dropna(subset=['percent_infected'])

# Ordenar o DataFrame pelo percentual de pessoas infectadas em ordem decrescente
df_sorted = df_grouped.sort_values(by='percent_infected', ascending=False)

# Criar uma figura com dois subplots (um abaixo do outro)
fig, axes = plt.subplots(2, 1, figsize=(10, 12))

# Plotagem de barras verticais
axes[0].bar(df_sorted['state'], df_sorted['percent_infected'], color='red')
axes[0].set_xlabel('Estado')
axes[0].set_ylabel('Percentual de Pessoas Infectadas')
axes[0].set_title('Percentual de Pessoas Infectadas por Estado (Barras Verticais)')

# Plotagem de barras horizontais
axes[1].barh(df_sorted['state'], df_sorted['percent_infected'], color='blue')
axes[1].set_xlabel('Percentual de Pessoas Infectadas')
axes[1].set_ylabel('Estado')
axes[1].set_title('Percentual de Pessoas Infectadas por Estado (Barras Horizontais)')

plt.tight_layout()  # Ajustar o espaçamento entre os subplots
plt.show()
```
